ASking a Question

What are the location with the fastest roller coaster

```python
df_subset['Location'].value_counts()
```

```python
axx = df_subset.query('Location != "Other"')\
    .groupby('Location')['speed_mph'] \
    .agg(['mean','count']) \
    .query('count > 10').sort_values('mean')\
    .plot(kind='barh' ,figsize=(15,5) ,title="AVg speed by Locations")
```
